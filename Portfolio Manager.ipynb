```python
class PortfolioManager:
    def __init__(self, totalPosition,numberOfPairs):
        self.totalPosition = totalPosition
        self.numberOfPairs = numberOfPairs
        # position per pair for each long or short
        self.posPerType = totalPosition/numberOfPairs/2
        self.pair = set()
        self.long_postion = {}
        self.short_position = {}
        self.realisedPnl = 0
        print("Position Per Pair per long/short",f"{self.posPerType:.1f}")
        
    def getRealisedPnl(self):
        return self.realisedPnl
        
    def getLongPosition(self):
        return self.realisedPnl
    
    def getLongPosition(self):
        return self.long_postion
    
    def getShortPosition(self):
        return self.short_position
    
    def getAllPosition(self):
        print("Long Position",self.long_postion)
        print("Short Position",self.short_position)
    def __getPairName(self,longTickerName,shortTickerName):
        return longTickerName+"-"+shortTickerName;
    
    def __getShareQuantity(self,price):
        return self.posPerType / price;
        
    def addPositon(self,longTickerName,shortTickerName,longPrice,shortPrice):
        pairName = self.__getPairName(longTickerName,shortTickerName)
        if len(self.pair) >= self.numberOfPairs:
            print("Pairs limit:",self.numberOfPairs)
            return
        if pairName in self.pair:
            print("Pair already Exist",pairName)
        else:
            self.pair.add(pairName)
            print("Adding position for",pairName)
            longQuantity = self.__getShareQuantity(longPrice)
            shortQuantiy = self.__getShareQuantity(shortPrice)

            self.__enterLongPosition(pairName,longQuantity,longPrice)
            self.__enterShortPosition(pairName,shortQuantiy,shortPrice)
        
        
    def closePosition(self,longTickerName,shortTickerName,closingLongPrice,closingShortPrice,tradingCost = 0):
        pairName = self.__getPairName(longTickerName,shortTickerName)
        if pairName not in self.pair:
            return
        msg = "Closing position for {}, trading cost :{} , closing long price {} closing short price".format(pairName,tradingCost)
        print(msg)
        longGrossPnl = self.__closeLongPosition(pairName,closingLongPrice)
        shortGrossPnl = self.__closeShortPosition(pairName,closingShortPrice)
        pnl = longGrossPnl + shortGrossPnl - tradingCost
        self.realisedPnl +=pnl
        self.pair.remove(pairName)
        return pnl;
        
    def __closeLongPosition(self,pairName,closingPrice):
        if pairName in self.long_postion:
            qty = self.long_postion[pairName]["quantity"]
            notional = self.long_postion[pairName]["price"] * qty
            market = qty * closingPrice;
            # market is sell ,notional is buy
            grossPnl = market - notional;
            del self.long_postion[pairName]
            return grossPnl
        else:
            print("Nothing to close long, pair does not exist",pairName)
            return 0;
    def __closeShortPosition(self,pairName,closingPrice):
        if pairName in self.short_position:
            qty = self.short_position[pairName]["quantity"]
            notional = self.short_position[pairName]["price"] * qty
            market = qty * closingPrice;
            # market is buy ,notional is sell
            grossPnl = notional - market;
            del self.short_position[pairName]
            return grossPnl
        else:
            print("Nothing to close short, pair does not exist",pairName)
            return 0;
    def __enterLongPosition(self,pairName,quantity,price):
        if pairName in self.long_postion:
            print("pair already exist",pairName)
        else:
            self.long_postion[pairName] = {}
            self.long_postion[pairName]["quantity"] = quantity
            self.long_postion[pairName]["price"] = price
            
    def __enterShortPosition(self,pairName,quantity,price):
        if pairName in self.short_position:
            print("pair already exist",pairName)
        else:
            self.short_position[pairName] = {}
            self.short_position[pairName]["quantity"] = quantity
            self.short_position[pairName]["price"] = price
```

```python

totalPosition = 100_000_000
noOfPairs = 100
tradingCost = 123
#create portfolio manager
portfolioManager = PortfolioManager(totalPosition,noOfPairs)
#add position
portfolioManager.addPositon("ABC","EFG",1.23,1.43)
#close position
portfolioManager.closePosition("ABC","EFG",1.33,1.33,tradingCost)
print("Realised Pnl" ,portfolioManager.getRealisedPnl())
# example pair breakage
#add position
portfolioManager.addPositon("WWW","ZZZ",1.55,2.43)
#close position
portfolioManager.closePosition("WWW","ZZZ",1.33,2.88)
print("Realised Pnl" ,portfolioManager.getRealisedPnl())

# test pairs limit
for i in range(102):
    index = str(i)
    portfolioManager.addPositon(index+"L",index+"S",1.23,1.43)
for i in range(102):
    index = str(i)
    portfolioManager.closePosition(index+"L",index+"S",1.33,1.33,tradingCost)
print("Realised Pnl" ,portfolioManager.getRealisedPnl())
```
